```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/airbnb_cleaned.csv")
df.shape
df.info()
```

```python
df.shape
```

```python
df["revenue"] = df["price"] * df["minimum nights"]
df["revenue"].describe()
```

```python
median_revenue = df["revenue"].median()
df["high_revenue"] = (df["revenue"] > median_revenue).astype(int)

df["high_revenue"].value_counts()
```

```python
df = pd.get_dummies(df, columns=[
    "room type",
    "neighbourhood group",
    "cancellation_policy"
], drop_first=True)
```

```python
df.to_csv("../data/airbnb_featured.csv", index=False)
```
